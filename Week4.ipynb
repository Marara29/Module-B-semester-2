```python
df_movies = pd.read_csv("https://raw.githubusercontent.com/Marara29/Boston-University-Activities/refs/heads/main/enhanced_box_office_data(2000-2024)u.csv?token=GHSAT0AAAAAAC57Y4SWIWXLN7ZW5LSFB4MG2CPSFSA")
```

```python
df_clean=df_movies.dropna()
categorical_columns = ["Genres", "Rating", "Original_Language", "Production_Countries"]
df = pd.get_dummies(df_clean, columns=categorical_columns, drop_first=False,dtype=int)
```

```python
df= df.drop(columns='Release Group')
```

```python
for col in df.columns:
    plt.figure(figsize=(16, 4))

    # Histogram
    plt.subplot(1, 4, 1)
    sns.histplot(df[col], bins=30, kde=False, color='skyblue')
    plt.title(f'Histogram: {col}')

    # KDE Plot
    plt.subplot(1, 4, 2)
    sns.kdeplot(df[col], fill=True, color='orange')
    plt.title(f'KDE Plot: {col}')

    # Violin Plot
    plt.subplot(1, 4, 3)
    sns.violinplot(x=df[col], color='lightgreen')
    plt.title(f'Violin Plot: {col}')

    plt.subplot(1, 4, 4)
    sns.swarmplot(x=df[col].sample(n=100, random_state=0), color='purple', size=2)
    plt.title(f'Swarm Plot: {col} (Sample of 100)')

    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['$Domestic'], color='blue', label='Domestic', bins=30, alpha=0.5)
sns.histplot(df['$Foreign'], color='red', label='Foreign', bins=30, alpha=0.5)
plt.title('Grouped Histogram: Domestic vs Foreign income')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

#  Draw a bar plot to indicate total counts of each categorical variable in a given column.Draw a bar plot to indicate total counts of each categorical variable in a given column.

```python
categorical_columns = ["Genres", "Rating", "Original_Language", "Production_Countries"]

for col in categorical_columns:
    plt.figure(figsize=(10, 5))
    

    counts = df_movies[col].value_counts().head(20)  
    
    counts.plot(kind='bar', color='skyblue')
    plt.title(f"Counts per Category in '{col}'")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
